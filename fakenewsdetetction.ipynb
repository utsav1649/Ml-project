```python
import re
```

```python

# Example cleaning function
def wordopt(text):
    # Convert into lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Remove digits
    text = re.sub(r'\d', '', text)

    # Remove newline characters
    text = re.sub(r'\n', ' ', text)

    return text
```

```python
def output_label(n):
    if n == 0:
        return "It is a Fake News"
    elif n == 1:
        return "It is a Genuine News"
```

```python
def manual_testing(news):
    # Put text into DataFrame
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)

    # Apply preprocessing
    new_def_test["text"] = new_def_test["text"].apply(wordopt)

    # Extract processed text
    new_x_test = new_def_test["text"]

    # Vectorize using fitted TF-IDF (very important!)
    new_xv_test = vectorization.transform(new_x_test)

    # Predictions from different models
    pred_lr = LR.predict(new_xv_test)
    pred_gbc = gbc.predict(new_xv_test)
    pred_rdf = RDF.predict(new_xv_test)

    return (
        f"\nLR Prediction: {output_label(pred_lr[0])}"
        f"\nGBC Prediction: {output_label(pred_gbc[0])}"
        f"\nRFC Prediction: {output_label(pred_rdf[0])}"
    )



```

```python
news_article = str(input("Enter a news article: "))
```

```python
print(manual_testing(news_article))
```
